```python
# Now we are having a dataset of an insuarance company and check which age group intrested in getting the insurance
```

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

#Matplotlib Inline command is a magic command that makes the plots generated by matplotlib
#how into the IPython shell that we are running and not in a separate output window.
```

```python
import requests
import io

url = "https://raw.githubusercontent.com/codebasics/py/master/ML/7_logistic_reg/insurance_data.csv"

download = requests.get(url).content

df = pd.read_csv(io.StringIO(download.decode('utf-8')))
```

```python
df
```

```python
plt.scatter(df.age , df.bought_insurance , marker = '+' , color = 'red')
```

```python
# from the graph we get to know if the person is young probability is less and if it is old then the probability is high
```

```python
from sklearn.model_selection import train_test_split
```

```python
# note use shift+tab for getting whole detail about the function
X_train, X_test, y_train, y_test=train_test_split(df[['age']],df.bought_insurance,test_size=0.1)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
# so all the people are old  , they will take the insurance
```

```python
clf.predict([[26]])
```

```python
clf.score(X_test,y_test)
```

```python
# for suggestion use the tab
clf.predict_proba(X_test)
```

```python
# [prob_not_buy , prob_buy] given above
```

# Multiclass Classification

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits #Each datapoint is a 8x8 image of a digit.
#Samples total -1797
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0] #filled array in 1d
```

```python
# As its an array of the filled part
```

```python
digits.images[0]
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5] # we will train our model by data and the target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
#let's check whether it meets the accuracy or not
```

```python
digits.target[89]
```

```python
model.predict([digits.data[89]])
```

```python
plt.matshow(digits.images[89])
```

```python
# now its only for one value what if i have to see the prediction made by the model
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_predicted)
```

```python
cm
```
